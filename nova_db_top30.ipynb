### Exercici 3
#### creació d'una nova base de dades únicament amb els 29 aeroports amb més tràfic

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('delayedflights.csv')
df.info()
```

```python
df1=df.Origin.value_counts()
type(df1)
```

```python
top=list()
n=0
for x,y in df1.items():
    if n<29:
        print(x)
        n=n+1
        top.append(x)
```

```python
mask = df['Origin'].isin(top)
```

```python
df_top = df[mask]
```

```python

mask1= df['Dest'].isin(top)
```

```python
df_top = df_top[mask1]
```
